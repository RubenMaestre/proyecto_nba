```python
# Me he dado cuenta que al pasarlo así se me queda una cosa muy fea, en formato AIRTABLE y no me gusta para hacer gráficas
# ni para leer los datos. Así que voy a trabajar con pandas para hacer el dataframe a mi gusto
# Voy a decirle que todas los datos del diccionario en el fields sean columnas y luego almacenar los datos
# Además eliminaré la columna id y la de tiempo que no me valen para nada para hacer gráficas

#Tenemos ya los datos en dataframe
df_datos_equipos_nba
df_datos_jugadores_nba

# Por lo que he visto en la documentación lo primero que hay qeu hacer es utilizar normalize a los fields
normalizar_equipos = pd.json_normalize(df_datos_equipos_nba['fields'])
normalizar_jugadores = pd.json_normalize(df_datos_jugadores_nba['fields'])

# Aquí para que lo tengamos en cuenta me volví loco hasta que encontré el fallo, había que poner una axis
# porque si no hacía las cosas en las filas y no en la columna... 
df_bueno_equipos = pd.concat([df_datos_equipos_nba.drop('fields', axis=1), normalizar_equipos], axis=1)
df_bueno_jugadores = pd.concat([df_datos_jugadores_nba.drop('fields', axis=1), normalizar_jugadores], axis=1)


```
